```python
def occupied(grid, lin, col):
    if lin < 0 or lin >= len(grid) or col < 0 or col >= len(grid[0]):
        return 0
    return int(grid[lin][col] == '#')

def step(grid):
    nb_lines = len(grid)
    nb_cols = len(grid[0])

    new_grid = []
    for lin in range(nb_lines):
        new_grid.append('')
        for col in range(nb_cols):
            if grid[lin][col] == '.':
                new_grid[lin] += '.'
                #print('.', end='')
            else:
                nb_neighbors = (occupied(grid, lin-1, col-1) + occupied(grid, lin-1, col) + occupied(grid, lin-1, col+1) +
                               occupied(grid, lin, col-1) + occupied(grid, lin, col+1) +
                               occupied(grid, lin+1, col-1) + occupied(grid, lin+1, col) + occupied(grid, lin+1, col+1))
                #print(nb_neighbors, end='')
                if grid[lin][col] == 'L' and nb_neighbors == 0:
                    new_grid[lin] += '#'
                elif grid[lin][col] == '#' and nb_neighbors >= 4:
                    new_grid[lin] += 'L'
                else:
                    new_grid[lin] += grid[lin][col]

    return new_grid
```

```python
def equals(grid1, grid2):
    for lin in range(len(grid1)):
        for col in range(len(grid1[0])):
            if grid1[lin][col] != grid2[lin][col]:
                return False
    return True
```

```python
def total_occupied(grid):
    total = 0
    for lin in range(len(grid)):
        for col in range(len(grid[0])):
            if grid[lin][col] == '#':
                total += 1
    return total
```

```python
cur_grid = input_grid
while not equals(step(cur_grid), cur_grid):
    cur_grid = step(cur_grid)
```

```python
def free_seat(grid, lin, col, d_lin, d_col): # is there a free seat in this direction
    nb_lin = len(grid)
    nb_col = len(grid[0])
    lin += d_lin
    col += d_col
    while not (lin < 0 or lin >= nb_lin or col < 0 or col >= nb_col):
        if grid[lin][col] == 'L':
            #print(f'found free seat at {lin} {col}')
            return 0
        elif grid[lin][col] == '#':
            #print(f'found occupied seat at {lin} {col}')
            return 1
        lin += d_lin
        col += d_col
    return 0
```

```python
def step_2(grid):
    nb_lines = len(grid)
    nb_cols = len(grid[0])

    new_grid = []
    for lin in range(nb_lines):
        new_grid.append('')
        for col in range(nb_cols):
            if grid[lin][col] == '.':
                new_grid[lin] += '.'
                #print('.', end='')
            else:
                nb_neighbors = (free_seat(grid, lin, col, -1, -1) + free_seat(grid, lin, col, -1, 0) +
                                free_seat(grid, lin, col, -1,  1) +
                                free_seat(grid, lin, col,  0, -1) + free_seat(grid, lin, col,  0, 1) +
                                free_seat(grid, lin, col,  1, -1) + free_seat(grid, lin, col,  1, 0) +
                                free_seat(grid, lin, col,  1,  1))
                #print(nb_neighbors, end='')
                if grid[lin][col] == 'L' and nb_neighbors == 0:
                    new_grid[lin] += '#'
                elif grid[lin][col] == '#' and nb_neighbors >= 5:
                    new_grid[lin] += 'L'
                else:
                    new_grid[lin] += grid[lin][col]

    return new_grid
```
