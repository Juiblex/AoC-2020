```python
real_input = [0, 13, 1, 8, 6, 15]
```

```python
def game(inp, nb_rounds):
    last_times = {}
    for t, n in enumerate(inp):
        #print(f'round {t+1}: speak {n}')
        last_times[n] = [t+1]

    last_spoken = inp[-1]

    time = len(inp) + 1

    while time <= nb_rounds:
        if time % 1000 == 0:
            print(f'round {time}')
        if last_spoken in last_times:
            if len(last_times[last_spoken]) == 1:
                new_spoken = 0
            else:
                new_spoken = last_times[last_spoken][-1] - last_times[last_spoken][-2]
        else:
            new_spoken = 0
        if new_spoken not in last_times:
            last_times[new_spoken] = []
        last_times[new_spoken].append(time)
        last_spoken = new_spoken
        time += 1
    
    return last_spoken
```

```python
game(real_input, 30000000)
```
